````python
df_merge = pd.merge(df_basics, df_ratings, left_on='tconst', right_on='tconst')
df_merge
````

````python
df_slim = df_merge
df_slim = df_slim.drop(df_slim[df_slim.startYear == '\\N'].index)
df_slim = df_slim[ (df_slim['titleType'].isin(['movie']) ) & (df_slim['isAdult'] == 0) ]
df_slim = df_slim.drop(['endYear', 'isAdult'], axis=1)

df_slim.insert(loc=2, column='year', value=df_slim['startYear'])
df_slim = df_slim.drop(['startYear'], axis=1)
df_slim
#dfy = df_merge[(df_merge['startYear'] == '\\N') & (df_merge['titleType'].isin(['movie','tvMovie']) & (df_merge['numVotes'] > 1))]
#dfy = df_merge[(df_merge['startYear'] == '\\N') & (df_merge['titleType'].isin(['movie','tvMovie']) & (df_merge['numVotes'] > 1))]
#dfy
````

````python
#df_slim['runtimeMinutes']
#temp = df_slim[ df_slim['runtimeMinutes'] == '\\N' ]['runtimeMinutes'].values.any()
df_slim = df_slim.replace(to_replace="\\N", value=np.NaN)
df_slim
````

````python
# Clean years, runtime columns
df_slim['year'] = df_slim['year'].fillna(0.0).astype(int)
df_slim['runtimeMinutes'] = df_slim['runtimeMinutes'].fillna(0.0).astype(int)

df_slim
````
