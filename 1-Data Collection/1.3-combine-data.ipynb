```python
import numpy as np
import pandas as pd
import csv
```

```python
list_of_dfs = []
for yr in range (1920,2020):
    file_path = f'rawdata/imdb_scrape_{yr}.tsv'
    temp_df = pd.read_csv(file_path, sep='\t', index_col=0)
    list_of_dfs.append(temp_df)
```

```python
# Combine all dfs
my_df = pd.concat(list_of_dfs, ignore_index=True)
```
